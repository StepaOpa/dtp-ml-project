```python
import geopandas as gpd
import pandas as pd
import json
from shapely.geometry import Point
```

```python
file_path = "C:/Users/lenovo/Desktop/URFU/DataAnalysis_3sem/DTP_GIBDD/sverdlovskaia-oblast.geojson"
with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

features = data['features']
df = pd.json_normalize(features)
```

Проеобразование координат в тип данных shapely.geometry.Point, проеобразование ячеек без значений в тип даных None, т.к. geopandas отказывается создавать geometry по NoneType значениям

```python
def make_point(coords):
    if coords is None:
        return None
    if isinstance(coords, list) and len(coords) == 2:
        try:
            lon, lat = float(coords[0]), float(coords[1])
            if -180 <= lon <= 180 and -90 <= lat <= 90:
                return Point(lon, lat)
        except (ValueError, TypeError):
            pass
    return None

geometry = df['geometry.coordinates'].apply(make_point)
```

```python

gdf = gpd.GeoDataFrame(df, crs='epsg:4326', geometry=geometry)
gdf.sample(3)
```
